```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/pacientes.json'
dados_paciente = pd.read_json(url)
dados_paciente
```

```python
url_2 = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/pacientes_2.json'
dados_paciente_2 = pd.read_json(url_2)
dados_paciente_2
```

```python
df_normalizado = pd.json_normalize(dados_paciente_2['Pacientes'])
```

```python
df_normalizado.to_json('historico_pacientes_normalizado.json', index=False)
```
