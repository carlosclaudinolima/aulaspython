
2.   **Análise exploratória dos dados**



Obter a média do valor de imóvel por tipo

```python
valores_medios_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor', ascending=True)
valores_medios_tipo.plot(kind='barh', title='Valor médio do aluguel por tipo de imóvel', color='purple')
```

Selecionndo os tipos de imóvel

```python
dados.Tipo.unique()
```

Criar um lista de tipos de imóveis comerciais

```python
comerciais = ['Conjunto Comercial/Sala',
              'Prédio Inteiro', 'Loja/Salão',
              'Galpão/Depósito/Armazém', 'Casa Comercial',
              'Terreno Padrão', 'Loja Shopping/ Ct Comercial',
              'Chácara', 'Loteamento/Condomínio','Sítio',
              'Pousada/Chalé', 'Hotel', 'Indústria']
```

Criar Datafrme que não incluia os imoóveis de tipos comerciais

```python
df = dados.query('Tipo not in @comerciais')
```

Plotando grafico das médias de valor de aluguel de imovel não comercial por tipo

```python
valores_medios_tipo_NC = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor', ascending=True)
valores_medios_tipo_NC.plot(kind='barh', title='Valor médio do aluguel por tipo de imóvel ( Não comercial )', color='purple')
```
