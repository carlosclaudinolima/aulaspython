1.   **Iniciando importação e reconhecimento dos dados**

Importar pandas

```python
import pandas as pd
```

Define a URL do dataset

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
```

Analisar o dataset

```python
pd.read_csv(url)
```

Arquivo usando ";" como separador. Passando parâmetro correto

```python
pd.read_csv(url, sep=";")
```

Ok, importando os dados e exibindo somente os primeiros registros

```python
dados = pd.read_csv(url, sep=";")
dados.head(10)
```

Exibindo os 10 ultimos registros


2.   **Análise exploratória dos dados**



Obter a média do valor de imóvel por tipo

```python
valores_medios_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor', ascending=True)
valores_medios_tipo.plot(kind='barh', title='Valor médio do aluguel por tipo de imóvel', color='purple')
```
