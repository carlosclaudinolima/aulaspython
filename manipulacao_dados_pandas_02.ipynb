```python
import pandas as pd
```

```python
url = "https://github.com/carlosclaudinolima/aulaspython/raw/refs/heads/main/dados/moveis_disponiveis.json"
```

```python
dados = pd.read_json(url)
```

```python
dados.info()
```

```python
dados['data'] = pd.to_datetime(dados['data'])
```

```python
dados.info()
```

```python
dados.head()
```

```python
dados['data'].dt.strftime('%Y-%m')
```

```python
subset = dados.groupby(dados['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
```
