```python
import pandas as pd
```

```python
url = '/content/sample_data/emissoes_gases 01.csv'
```

```python
emissao_gases = pd.read_csv(url)
```

```python
emissao_gases.info()
```

```python
colunas_info = list(emissao_gases.loc[:, 'Nível 1 - Setor':'Produto'].columns)
colunas_info
```

```python
colunas_emissao = list(emissao_gases.loc[:, '1970':'2021'])
colunas_emissao
```

```python
emissao_gases.melt(id_vars= colunas_info, value_vars= colunas_emissao, var_name='Ano', value_name='Emissão')
```

```python
emissoes_por_ano = emissao_gases.melt(id_vars= colunas_info, value_vars= colunas_emissao, var_name='Ano', value_name='Emissão')
```

```python
emissao_por_gas = emissoes_por_ano.groupby('Gás')[['Emissão']].sum().sort_values('Emissão', ascending=False)
emissao_por_gas
```

```python
emissao_por_gas.plot(kind='barh', figsize=(10, 6))
```
