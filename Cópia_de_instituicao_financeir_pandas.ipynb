```python
 from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
eng = create_engine('sqlite:///:memory:')
```

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
dados.to_sql('clientes', eng, index=False)
```

```python
inspecor = inspect(eng)
```

```python
print(inspecor.get_table_names())
```

```python
sql = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
empregados = pd.read_sql(sql, eng)
```

```python
empregados
```

```python
empregados.to_sql('empregados', con=eng, index=False, if_exists='replace')  #Replace, taco por cima mesmo,sou mau, hahahahaa
```

```python
inspecor = inspect(eng)
inspecor.get_table_names()
```

```python
pd.read_sql_table('empregados', eng, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```
